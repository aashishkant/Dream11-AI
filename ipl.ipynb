### Fantasy Points System

This code defines a compact system for calculating fantasy points in cricket based on player performances. Key actions are assigned points as follows:

**Batting:**
- **Run:** +1 point per run
- **Boundary (bFour):** +1 point per boundary (four runs)
- **Six (bSix):** +2 points per six
- **30 Runs:** +4 points
- **Half-century:** +8 points
- **Century:** +16 points
- **Duck:** -2 points
- **Strike Rate (sr):** Variable points based on different strike rate thresholds

**Bowling:**
- **Wicket:** +25 points per wicket
- **LBW/Bowled:** +8 points
- **Wicket Hauls:** Additional points for 3, 4, or 5 wickets
- **Maiden:** +12 points
- **Economy Rate (rpo):** Variable points based on different economy rate thresholds

**Fielding:**
- **Catch:** +8 points per catch
- **Multiple Catches:** Bonus points for 3 or more catches
- **Stumping:** +12 points
- **Run Outs:** +12 points for direct and +6 points for indirect

This scoring system is used to evaluate player performances in fantasy cricket, influencing team selection and strategy.

## IPL Fantasy Cricket Team Prediction

This Python code predicts fantasy cricket team players based on their performances. It defines a function `get_players()` to fetch players from each team and evaluate their fantasy points. The function considers various factors like runs scored, boundaries, wickets taken, and recent performance to calculate the final fantasy points for each player.


```python
import pandas as pd

# Define the function to get players from each team with their fantasy points
def get_players(team1, team2, team1_fp):
    fantasy_team_players = []

    for i in range(len(team1)):
        # Filtering unique match IDs for the current player in team1
        unq_ids = byb[byb["batsman"] == team1[i]]['id'].unique()
        mathces_played = len(unq_ids)
        
        bbr = []
        for x in unq_ids:
            # Calculating total runs scored by the player in each match
            bat_run = sum(byb[(byb["batsman"] == team1[i]) & (byb['id'] == x)]['batsman_runs'])
            bbr.append(bat_run)

        # Calculating the number of 30s, 50s, and 100s
        r30, r50, r100 = 0, 0, 0
        for m in bbr:
            if m >= 100:
                r100 += 1
            elif m >= 50:
                r50 += 1
            elif m >= 30:
                r30 += 1
        
        # Calculating extra points for batsmen
        try:
            catches = len(byb[(byb['fielder'] == team1[i]) & (byb['dismissal_kind'] == 'caught')]) / mathces_played
            run_outs = len(byb[(byb['fielder'] == team1[i]) & (byb['dismissal_kind'] == 'run out')]) / mathces_played
            extra_points = r30 / mathces_played * Batsman_points['30Runs'] + r50 / mathces_played * Batsman_points['Half_century'] + r100 / mathces_played * Batsman_points['Century'] + catches * Fielding_points['Catch'] + run_outs * Fielding_points['RunOutInd']
        except:
            catches, run_outs, extra_points = 0, 0, 0

        # Calculating extra points for bowlers
        wickets_taken = []
        for x in unq_ids:
            twx = sum(byb[(byb["bowler"] == team1[i]) & (byb['id'] == x)]['is_wicket'])
            wickets_taken.append(twx)

        w3, w4, w5 = 0, 0, 0
        for z in wickets_taken:
            if z >= 5:
                w5 += 1
            elif z >= 4:
                w4 += 1
            elif z >= 3:
                w3 += 1
        try:
            lbws = len((byb[(byb['bowler'] == team1[i]) & (byb['dismissal_kind'] == 'lbw')])) / mathces_played
            bowled = len((byb[(byb['bowler'] == team1[i]) & (byb['dismissal_kind'] == 'bowled')])) / mathces_played
            wexp = w3 / mathces_played * Bowling_points['3W'] + w4 / mathces_played * Bowling_points['4W'] + w5 / mathces_played * Bowling_points['5W'] + lbws * Bowling_points['LBW_Bowled'] + bowled * Bowling_points['LBW_Bowled']
        except:
            lbws, bowled, wexp = 0, 0, 0

        ffp = []
        for j in range(len(team2)):
            # Filtering interactions between the current batsman in team1 and bowlers in team2
            bat_vs_bowl = byb[(byb["batsman"] == team1[i]) & (byb["bowler"] == team2[j])]
            bowls_played = len(bat_vs_bowl)
            runs_scored = sum(bat_vs_bowl['batsman_runs'])
            fours = len(bat_vs_bowl[bat_vs_bowl['batsman_runs'] == 4])
            sixes = len(bat_vs_bowl[bat_vs_bowl['batsman_runs'] == 6])
            wicket = sum(bat_vs_bowl['is_wicket'])
            if bowls_played <= 6 * 10 and wicket >= 5:
                penalty = -16
            elif bowls_played <= 6 * 8 and wicket >= 4:
                penalty = -8
            elif bowls_played <= 6 * 6 and wicket >= 3:
                penalty = -4
            else:
                penalty = 0

            try:
                strike_rate = int(runs_scored / bowls_played * 100)
            except:
                strike_rate = 'NA'
            if bowls_played >= 10 and strike_rate != 'NA':
                if strike_rate >= 170:
                    pass  # Add your logic here
                elif strike_rate >= 150:
                    pass  # Add your logic here

            bowl_vs_bat = byb[(byb["bowler"] == team1[i]) & (byb["batsman"] == team2[j])]
            wicket_took = sum(bowl_vs_bat['is_wicket'])
            fantasy_points1 = runs_scored + fours * Batsman_points['bFour'] + sixes * Batsman_points['bSix'] - wicket * Bowling_points['Wicket'] + wicket_took * Bowling_points['Wicket'] + penalty
            ffp.append(fantasy_points1)
            print(team1[i], "against", team2[j], "Runs", runs_scored,
                  "bowls", bowls_played, "strike rate", strike_rate,
                  'Out', wicket, 'times', "Fours", fours, "Sixes", sixes, "fantasy points", fantasy_points1)
        sum_ffp = sum(ffp)
        if team1_fp[team1[i]] > 0:
            recent_performance_points = np.log(team1_fp[team1[i]])
        elif team1_fp[team1[i]] < 0:
            recent_performance_points = -np.log(abs(team1_fp[team1[i]]))
        else:
            recent_performance_points = 0

        # Calculating final fantasy points
        weight1 = 0.5
        weight2 = 1 - weight1
        final_fantasy_point = (sum_ffp + extra_points + wexp) * weight1 + recent_performance_points * weight2
        final_fantasy_point = round(final_fantasy_point, 2)
        fantasy_team_players.append((final_fantasy_point, team1[i]))

    fantasy_team_players.sort(reverse=True)
    return fantasy_team_players

```

```python
# Helper function to print player details
def print_player_details(player, team, opposing_team, team_squad, match_data):
    # Filter match data for the player against the opposing team
    bat_vs_bowl = match_data[(match_data["batsman"] == player) & (match_data["bowler"].isin(opposing_team))]
    bowls_played = len(bat_vs_bowl)
    runs_scored = bat_vs_bowl["batsman_runs"].sum()
    fours = len(bat_vs_bowl[bat_vs_bowl['batsman_runs'] == 4])
    sixes = len(bat_vs_bowl[bat_vs_bowl['batsman_runs'] == 6])
    wickets = bat_vs_bowl["is_wicket"].sum()

    # Filter match data for the player as a bowler against the opposing team
    bowl_vs_bat = match_data[(match_data["bowler"] == player) & (match_data["batsman"].isin(opposing_team))]
    wickets_taken = bowl_vs_bat["is_wicket"].sum()

    print(f"{player} against {', '.join(opposing_team)}:")
    print(f"Runs scored: {runs_scored}")
    print(f"Balls faced/bowled: {bowls_played}")
    print(f"Fours: {fours}")
    print(f"Sixes: {sixes}")
    print(f"Wickets taken: {wickets_taken}")
    print(f"Wickets lost: {wickets}")

    try:
        strike_rate = runs_scored / bowls_played * 100
        print(f"Strike rate: {strike_rate:.2f}")
    except ZeroDivisionError:
        print("Strike rate: N/A (No balls faced/bowled)")

    try:
        recent_performance = team_squad[player]
        print(f"Recent performance points: {recent_performance}")
    except KeyError:
        print(f"No recent performance data found for {player}")

    print()

# Example usage
team1 = lsg
team2 = pbks

print("Team 1 (LSG) player details:")
for player in team1:
    print_player_details(player, team1, team2, lsg_fp, byb)

print("Team 2 (PBKS) player details:")
for player in team2:
    print_player_details(player, team2, team1, pbks_fp, byb)
```

```python
# Define the function to get the final predicted team
def get_final_predicted_team(team1, team2, team1_fp, team2_fp):
    # Get players for each team
    t1 = get_players(team1, team2, team1_fp)
    t2 = get_players(team2, team1, team2_fp)

    # Combine players from both teams
    t3 = t1 + t2

    # Sort the combined list by fantasy points
    t3.sort(reverse=True)

    # Choose one captain from each team
    captain_team1 = max(t1, key=lambda x: x[0])[1]  # Captain from team 1
    captain_team2 = max(t2, key=lambda x: x[0])[1]  # Captain from team 2

    # Select top 10 players and include the captains
    result = [player for _, player in t3[:10]]
    result.append(captain_team1)  # Add captain from team 1
    result.append(captain_team2)  # Add captain from team 2

    # Create a DataFrame with the final predicted team
    final_team = pd.DataFrame(result, columns=['Player'])

    return final_team

# Define the function to choose captain and vice-captain
def choose_captain_and_vice_captain(final_team):
    # Sort the final team DataFrame by fantasy points in descending order
    final_team.sort_values(by='Fantasy Points', ascending=False, inplace=True)
    
    # Choose the player with the highest fantasy points as captain
    captain = final_team.iloc[0]['Player']
    
    # Choose the player with the second highest fantasy points as vice-captain
    vice_captain = final_team.iloc[1]['Player']
    
    return captain, vice_captain

# Specify the teams
team1 = rr
team2 = kkr

# Get the final predicted team
final_predicted_team = get_final_predicted_team(team1, team2, rr_fp, kkr_fp)

# Assign fantasy points to the final predicted team (Replace this with your logic)
final_predicted_team['Fantasy Points'] = np.random.randint(50, 100, len(final_predicted_team))

# Choose captain and vice-captain
captain, vice_captain = choose_captain_and_vice_captain(final_predicted_team)

# Print the final predicted team along with captain and vice-captain
print('\nFinal Predicted Team:\n', final_predicted_team)
print('\nCongratulations! The captain of your team is', captain)
print('Congratulations! The vice-captain of your team is', vice_captain)
```
